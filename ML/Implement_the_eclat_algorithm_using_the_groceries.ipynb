##Importing the libraries

```python

import pandas as pd
import numpy as np
from pyECLAT import ECLAT
```

##Data Preprocessing

```python
df = pd.read_csv('/content/groceries.csv',header = None)
df
```

```python
df.info()
```

##replacing empty value with 0.

```python
df.fillna(0,inplace=True)
df.head()
```

```python
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,32) if str(df.values[i,j])!='0'])
transactions[0]
```

```python
data = pd.DataFrame(transactions)
data
```

```python
min_n_products = 2

min_support = 7/len(transactions)

max_length = max([len(x) for x in transactions])
```

```python
my_eclat = ECLAT(data=data, verbose=True)

rule_indices, rule_supports = my_eclat.fit(min_support=min_support, min_combination=min_n_products, max_combination=2)
```
