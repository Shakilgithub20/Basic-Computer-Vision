## Import Libraries

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

## Import Dataset

```python
img = cv2.imread('/content/Old-Newspapers.jpg')
cv2_imshow(img)
height,width = img.shape[:2]
```

## Affine transform

```python
Tx,Ty =height/4,width/4
T = np.float32([[1,0,Tx],[0,1,Ty]])
result = cv2.warpAffine(img,T,(width,height))
cv2_imshow(result)
```

## Rotate

```python
height,width = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((width/2,height/2),180,1)
result = cv2.warpAffine(img,rotation_matrix,(width,height))
cv2_imshow(result)
```

## Scaling

```python
img_scaled = cv2.resize(img,None,fx=2,fy=2)
cv2_imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img,(200,200),interpolation = cv2.INTER_CUBIC)
cv2_imshow(img_scaled)
```

```python
height,width = img.shape[:2]
start_row,start_col = int(height*0.25),int(width*0.25)
end_row,end_col = int(height*0.55),int(width*0.55)
crop_img = img[start_row:end_row,start_col:end_col]
cv2_imshow(img)
cv2_imshow(crop_img)
```
